```python
import sqlite3
import pandas as pd

# Create a temporary SQLite database
db = sqlite3.connect('booking.db')
cursor = db.cursor()
```

# Table1

```python
# Create a table with appropriate column names, and use double quotes for column names with spaces
cursor.execute('''
CREATE TABLE table1 (Sno INTEGER, 'Hotel Name' VARCHAR, City VARCHAR, Location VARCHAR,
'Number of Reviews' INTEGER,
'User Ratings' INTEGER,
'Star Rating' INTEGER,
'Review Score' VARCHAR
)''')

# Read data from a CSV file using pandas
data_tab = pd.read_csv('E:\Surendra Data\DATA Scientist Course Detail\Internship\Capstone Project\Data\Data1.csv')  # Replace with the path to your CSV file

# Insert data into the SQLite table
data_tab.to_sql('table1', db, if_exists='append', index=False)
```

1) Write a query to give details of hotels city is Sydney and ratings is more than 5.

```python
# SQL query to retrieve details of hotels in Sydney with ratings greater than 5
results = cursor.execute("SELECT * FROM table1 WHERE City='Sydney' AND 'User Ratings' > 5")
for i in results:
    print(i)
```

2) Write a query to find the location with the most reviews.

```python
# query to find the location with the most reviews.
results = cursor.execute("SELECT Location, MAX('Number of Reviews') AS Max_Reviews FROM table1")
for i in results:
    print(i)
```

3) Write a query that will retrieve all rows and sort them in descending order of user ratings.

```python
results = cursor.execute("SELECT * FROM table1 ORDER BY 'User Ratings' DESC ")
for i in results:
    print(i)
```

4) Write a query that retrieves rows where the star rating is greater than or equal to 4.0 and sorts them in descending order of review score.

```python
results = cursor.execute("SELECT * FROM table1 WHERE 'Star Rating' >=4 ORDER BY 'Star Rating' DESC")
for i in results:
    print(i)
```

5) Write a query to count the number of hotels in each city. 

```python
results = cursor.execute("SELECT City, COUNT('Hotel Name') FROM table1 GROUP BY City")
for i in results:
    print(i)
```

# Table 3

```python
# Delete table if already exists
delete = "DROP TABLE table3"
cursor.execute(delete)
```

```python
# Create and insert data into Data3 table
cursor.execute('''CREATE TABLE table3 ( Sno INTEGER, 'Hotel Name' VARCHAR, 'Room Type' VARCHAR, Doublebed VARCHAR, Singlebed VARCHAR, Prices INTEGER
)''')

# Read data from a CSV file using pandas
data_tab3 = pd.read_csv('E:\Surendra Data\DATA Scientist Course Detail\Internship\Capstone Project\Data\Data3.csv')  # Replace with the path to your CSV file



# Insert data into the SQLite table
data_tab3.to_sql('table3', db, if_exists='append', index=False)
```

1) Write SQL query to retrieve hotels that offer double bed rooms and sort them by price in descending order.

```python
results = cursor.execute("SELECT DISTINCT 'Hotel Name' FROM table3 WHERE Doublebed='Yes' ORDER BY Prices DESC")
for i in results:
    print(i)
```

2) Write a SQL query to calculate the average price for each room type.

```python
results = cursor.execute("SELECT 'Room Type', AVG(Prices) AS avg_price FROM table3 GROUP BY 'Room Type'")
for i in results:
    print(i)
```

3) Write a query to find details of hotels with user ratings above 5 as a threshold and that offer free WiFi. (Use table 1 and table 2)

```python
results = cursor.execute("SELECT 'Hotel Name' FROM table1 WHERE 'User Ratings' > 5 AND 'Hotel Name' IN (SELECT 'Hotel Name' FROM table2 WHERE Freewifi = 'Yes')")
for i in results:
    print(i)
```

4) Write a query to find hotels that offer both family rooms and non-smoking rooms (Use table 1 and Table 2)

```python
results = cursor.execute("SELECT 'Hotel Name' FROM table1 WHERE 'Hotel Name' IN (SELECT 'Hotel Name' FROM table2 WHERE 'Family Rooms' = 'Yes' AND 'Non-Smoking Rooms' = 'Yes')")
for i in results:
    print(i)
```

5) Write a query to calculate the average price for each room type (Use table2 and Table 3) 

```python
results = cursor.execute("SELECT table3.'Room Type', AVG(table3.Prices) AS 'Average Price' FROM Table2 AS table2 JOIN Table3 AS table3 ON table2.'Sno' = table3.'Sno' GROUP BY table3.'Room Type';")
for i in results:
    print(i)
```
